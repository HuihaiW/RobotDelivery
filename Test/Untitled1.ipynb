```python
import pandas as pd
```

```python
df = pd.read_csv('../Data/intersections.csv')
df_road = pd.read_csv(r'../Data/road_segments.csv')
```

```python
lat_list = df['Lat'].values.tolist()
lon_list = df['Lon'].values.tolist()
lat_lon_list = []
for i in range(len(lat_list)):
    lat_lon_list.append(str(lat_list[i]) + str(lon_list[i]))
df['Lat_Lon'] = lat_lon_list
```

```python
connection_list = []
for lat_lon in set(lat_lon_list):
    new_df = df[df['Lat_Lon'] == lat_lon]
    connection = new_df['FID_Georgi'].values.tolist()
    connection_list.append(connection)
```

```python
adj_m = []
for connects in connection_list:
    for i in range(len(connects)):
        if i > len(connects)-2:
            continue
        else:
            for j in range(i + 1, len(connects)):
                adj_m.append([connects[i], connects[j]])
```
